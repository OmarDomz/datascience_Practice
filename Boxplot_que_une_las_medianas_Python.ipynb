##Ejemplos de un boxplot con una línea que una las medianas.

```python
import pandas as pd
import matplotlib.pyplot as plt

# Cargar datos desde archivo csv
ruta = 'https://raw.githubusercontent.com/gchavez83/Analitica/main/PowerBI_Latino.csv'
datos = pd.read_csv(ruta)

# Crear un diccionario con las categorías de carrera y sus valores de edad
datos_por_carrera = {}
for carrera in datos['Carrera'].unique():
    datos_por_carrera[carrera] = datos.loc[datos['Carrera'] == carrera, 'Edad'].values

# Crear un gráfico de cajas y bigotes con la función boxplot de matplotlib
fig, ax = plt.subplots()
ax.boxplot(datos_por_carrera.values())

# Agregar una línea que conecte las medianas de las categorías
mediana_por_carrera = {carrera: datos.loc[datos['Carrera'] == carrera, 'Edad'].median() for carrera in datos['Carrera'].unique()}
mediana_x = range(1, len(mediana_por_carrera) + 1)
mediana_y = list(mediana_por_carrera.values())
ax.plot(mediana_x, mediana_y, color='red', linewidth=2, linestyle='--', marker='o', markersize=8)

# Configurar el eje x para que muestre las categorías de carrera
ax.set_xticklabels(datos_por_carrera.keys())

# Agregar etiquetas de ejes y título
ax.set_xlabel('Carrera')
ax.set_ylabel('Edad')
ax.set_title('Gráfico de cajas y bigotes con medianas conectadas')
plt.show()

```

# Ejemplo 2

```python
# Importar las bibliotecas necesarias
import pandas as pd
import matplotlib.pyplot as plt

# Cargar los datos desde el archivo CSV
ruta = "https://raw.githubusercontent.com/gchavez83/Analitica/main/BoxPlot_Lines.csv"
datos = pd.read_csv(ruta)

# Crear el gráfico de cajas y bigotes con matplotlib
plt.figure(figsize=(8,6))
bp = datos.boxplot(column='Valor', by='Indicador', figsize=(8,6), grid=False)

# Agregar las líneas que unen las medianas
medians = datos.groupby(['Indicador'])['Valor'].median()
for i, median in enumerate(medians):
    plt.plot([i+1.1, i+0.9], [median, median], color='red', linewidth=2, zorder=10)

# Agregar la línea que une las medianas
x_coords = [i+1 for i in range(len(medians))]
y_coords = medians.values
plt.plot(x_coords, y_coords, color='blue', linewidth=2, zorder=10)

# Ajustar las etiquetas de los ejes y el título
plt.xlabel("Indicador")
plt.ylabel("Valor")
plt.title("Boxplot con medianas y línea unida por categoría")

# Mostrar el gráfico
plt.show()

```
