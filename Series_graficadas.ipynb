```python
import random
from collections import Counter
import plotly.graph_objs as go
from plotly.subplots import make_subplots

N = 6
sets_repetidos = []
num_iteraciones = 0
frecuencias_totales = Counter()
sets_deseados_por_Serie = 10

# Crear una figura separada para cada serie
figs = []
for i in range(N):
    sets_generados = []
    numeros_orden = 0
    sets_repetidos_serie = []
    num_iteraciones_serie = 0

    while len(sets_repetidos_serie) < sets_deseados_por_Serie:
        nuevo_set = set(random.sample(range(0, 47), 6))
        numeros_orden += 1
        if nuevo_set in sets_generados:
            sets_repetidos_serie.append(nuevo_set)
        else:
            sets_generados.append(nuevo_set)
            num_iteraciones += 1
            num_iteraciones_serie += 1

    sets_repetidos.append(sets_repetidos_serie)
    print(f"Los primeros 5 sets repetidos de la serie {i+1} con {num_iteraciones_serie} iteraciones, son:")
    for j in range(sets_deseados_por_Serie):
        print(f"Set {j+1}: {sets_repetidos_serie[j]}")
    print()

    # Contar la cantidad de veces que se repiten los números en la serie actual
    frecuencias = Counter([n for s in sets_generados for n in s])
    frecuencias_totales += frecuencias
    print(f"La frecuencia de los números en la serie {i+1} es: {frecuencias}\n")

    # Crear un gráfico de puntos de la frecuencia de los números en la serie actual
    fig = make_subplots(rows=1, cols=1)
    fig.add_trace(go.Scatter(x=list(frecuencias.keys()), y=list(frecuencias.values()), mode='markers'), row=1, col=1)
    fig.update_layout(title=f'Frecuencia de los números en la serie {i+1}', xaxis_title='Número', yaxis_title='Frecuencia')
    figs.append(fig)

# Mostrar todas las figuras generadas
for fig in figs:
    fig.show()
```
